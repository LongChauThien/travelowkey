# BUS

```python
def merge_element(arr):
  result1 = []
  result2 = []
  temp = []
  for element in arr:
    temp.append(element)
    if element == "City" or element == "Province":
      if (result1 == []):
        result1 = temp
      else:
        result2 = temp
      temp = []
  return (' '.join(result1)).strip(), (' '.join(result2)).strip()

def findArrivalTime(arr):
  PickUpPoint = ""
  DesPoint = ""
  temp = []
  ArrivalTime = ""
  for i in arr:
    if ":" in i:
      ArrivalTime = i
      PickUpPoint = (" ".join(temp)).strip()
      temp = []
      continue
    temp.append(i)
  DesPoint = (" ".join(temp)).strip()
  return PickUpPoint, DesPoint, ArrivalTime


f = open('bus.json', encoding="utf8")

data = json.load(f)

dict_new = {'DeparturePlace':"",
         "ArrivalPlace":"",
         "NumberOfCus":"",
         "DepartureTime":"",
         "Name":"",
         "TypeOfVe":"",
         "DepartureTime":"",
         "ArrivalTime":"",
         "PickUpPoint":"",
         "DesPoint":"",
         "TravelTime":"",
         "Price":""}
data_dict = []
for i in data['bus-ticket']:
    dict_ = {'DeparturePlace':"",
         "ArrivalPlace":"",
         "NumberOfCus":"",
         "DepartureTime":"",
         "Name":"",
         "TypeOfVe":"",
         "DepartureTime":"",
         "ArrivalTime":"",
         "PickUpPoint":"",
         "DesPoint":"",
         "TravelTime":"",
         "Price":""}
    string_info = i["info"]
    split_string = re.split("Thứ", string_info)
    string = split_string[0].strip()
    arr_string  = string.split()
    DeparturePlace, ArrivalPlace = merge_element(arr_string)
    dict_["DeparturePlace"] = DeparturePlace
    dict_["ArrivalPlace"] = ArrivalPlace
    dict_["NumberOfCus"] = split_string[1][-10:].strip()
    dict_["DepartureTime"] = ("Thứ"  + split_string[1][0:-10]).strip()
    dict_["Name"] = i["bus"]
    dict_["TypeOfVe"] = i["num"]
    dict_["Price"] = i["price"].split()[0]
    string_detail = i["detail"]
    string_detail = string_detail.split()
    dict_["DepartureTime"] = string_detail[0]
    dict_["TravelTime"] = ' '.join(string_detail[-2:])
    dict_["PickUpPoint"], dict_["DesPoint"], dict_["ArrivalTime"] = findArrivalTime(string_detail[1:-2])
    data_dict.append(dict_)
# Closing file
f.close()
```

# FLIGHT

```python
f = open('flights.json', encoding="utf8")

data = json.load(f)
data_dict = []
```

```python
for i in data['flight-ticket']:
    dict_ = {'DeparturePlace':"",
         "ArrivalPlace":"",
         "NumberOfPass":"",
         "DepartureDate":"",
         "SeatClass":"",
         "Name":"",
         "DepartureTime":"",
         "ArrivalTime":"",
         "TravelTime":"",
         "Stop/Direct":"",
         "+d":"",
         "CodeDepartureAirPort":"",
         "CodeArrivalAirPort":"",
         "Price":""}
    string_info = i["info"]
    split_string = re.split("\\|", string_info)
    string_ = split_string[0].strip()
    string_ = re.split("→", string_)
    dict_["DeparturePlace"] = string_[0].strip()
    flag = "Thứ" in string_[1].strip()
    string_ =  re.split("Thứ|CN", string_[1].strip())
    dict_["ArrivalPlace"] = string_[0].strip()
    if flag == True:
      dict_["DepartureDate"] = "Thứ " + string_[1].strip()
    else:
      dict_["DepartureDate"] = "CN" + string_[1].strip()
    dict_["NumberOfPass"] = split_string[1].strip()
    dict_["SeatClass"] = split_string[2].strip()
    dict_["Name"] = i["airline"].strip()
    dict_["Price"] = i["price"].split()[0].strip()
    string_detail = i["detail"]
    string_detail = string_detail.split()
    dict_["DepartureTime"] = string_detail[0].strip()
    dict_["CodeDepartureAirPort"] = string_detail[1].strip()
    if "+" in string_detail[-1]:
      dict_["+d"] = string_detail[-1].strip()
      dict_["CodeArrivalAirPort"] = string_detail[-2].strip()
      dict_["ArrivalTime"] = string_detail[-3].strip()
      dict_["Stop/Direct"] = string_detail[-5] + " " + string_detail[-4]
      if string_detail[-7] !=  string_detail[1]:
        dict_["TravelTime"] = string_detail[-7] + " " + string_detail[-6]
      else:
        dict_["TravelTime"] = string_detail[-6]
    else:
      dict_["CodeArrivalAirPort"] = string_detail[-1].strip()
      dict_["ArrivalTime"] = string_detail[-2].strip()
      dict_["Stop/Direct"] = string_detail[-4] + " " + string_detail[-3]
      if string_detail[-6] !=  string_detail[1]:
        dict_["TravelTime"] = string_detail[-6] + " " + string_detail[-5]
      else:
        dict_["TravelTime"] = string_detail[-5]
    data_dict.append(dict_)
# Closing file
f.close()
```

# ROOM

```python
f = open('rooms.json', encoding="utf8")

data = json.load(f)
data_dict = []
```

```python
for i in data['rooms']:
    dict_ = {'Name':"",
         "Max-numberOfCustomer":"",
             "Service":"",
             "Price":""}
    dict_["Name"] = i["name"]
    dict_["Max-numberOfCustomer"] = i["max_people"].split()[0]
    dict_["Service"] = i["service"].replace("\n", ", ")
    dict_["Price"] = i["price"].split()[0]
    data_dict.append(dict_)
# Closing file
f.close()
```

# TAXI

```python
f = open('taxi.json', encoding="utf8")

data = json.load(f)
data_dict = []
```

```python
for i in data['taxi']:
    dict_ = {'TypeOfCar':"",
             "DepartureTime":"",
             "DepartureDate":"",
             "ArrivalTime":"",
             "ArrivalDate":"",
             "PickUpPoint":"",
             "Luggage":"",
             "NumberOfSeat":"",
             "Price":""}
    dict_["TypeOfCar"] = i["taxi"]
    dict_["NumberOfSeat"] = i["detail"].split()[-2]
    dict_["Luggage"] = i["detail"].split()[0]
    dict_["Price"] = i["price"].split()[0]
    temp_string = i["info"]
    temp_string = re.split("•", temp_string)
    dict_["PickUpPoint"] = temp_string[0].strip()
    temp_string_ = re.split("-", temp_string[1])
    dict_["DepartureTime"] = temp_string_[0].split()[-1]
    dict_["DepartureDate"] = (" ".join(temp_string_[0].split()[:-1])).strip()
    dict_["ArrivalTime"] = temp_string_[1].split()[-1]
    dict_["ArrivalDate"] = (" ".join(temp_string_[1].split()[:-1])).strip()
    data_dict.append(dict_)
# Closing file
f.close()
```
